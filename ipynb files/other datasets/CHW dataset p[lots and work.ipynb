```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
chw = pd.read_excel("CHW.xlsx")
```

```python
chw
```

```python
chw = chw.ffill()
chw['Column23'] = abs(chw['Column3'] - chw['Column4'])
chw['Column24'] = chw['Column5']
chw['Column25'] = chw['Column6']
chw['Column26'] = chw['Column7']
chw
```

```python
chw.columns.values
```

```python
chw.to_excel("chw_filled.xlsx")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
df1 = chw.copy()

df1 = df1.drop(['Column5','Column6','Column7','Column8','Column9','Column10','Column11','Column12','Column13','Column14','Column15','Column16','Column17'],1)
df1 = df1.drop(['Column18','Column19','Column20','Column21','Column1','Column2'],1)
df1.rename(columns={'Column3': 'CW return temp','Column22':'Outside Temperature', 'Column23': 'Delta Temp', 'Column4':'Supply Temperature'}, inplace = True)
df1.rename(columns={'Column24': 'Chiller 1 Capacity', 'Column25': 'Chiller 2 Capacity', 'Column26': 'Chiller 3 Capacity'}, inplace = True)
df1.head()
```

```python
def plotdual(x_col,y_col):
    df1.plot.scatter(x=x_col, y=y_col)
    plt.savefig(''.join([x_col,y_col,',png']))
    plt.hexbin(x=df1[x_col], y =df1[y_col], gridsize = 9)
    plt.savefig(''.join([x_col,y_col,'_','hexplot']))
    
def bivariate(df1):
    df = pd.scatter_matrix(df1, alpha = 0.2, figsize = (50,50), diagonal = 'kde')
    plt.savefig(''.join(['bivariate',df1])
```
